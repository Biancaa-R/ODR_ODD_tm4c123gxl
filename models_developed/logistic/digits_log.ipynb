1. One-vs-Rest (OvR) (One-vs-All)
This is the default strategy for multiclass classification in Scikit-learn's Logistic Regression when the number of classes is greater than two.

In the One-vs-Rest (OvR) approach, a separate binary classifier is trained for each class. Each classifier learns to distinguish between one specific class and all the others. For example, if there are 10 classes (e.g., digits 0-9), the model will train 10 classifiers:

Classifier 1: Identifies whether the sample is class 0 or not (0 vs. non-0).
Classifier 2: Identifies whether the sample is class 1 or not (1 vs. non-1).
Classifier 3: Identifies whether the sample is class 2 or not (2 vs. non-2).
...
Classifier 10: Identifies whether the sample is class 9 or not (9 vs. non-9).
At prediction time, each classifier gives a probability that the sample belongs to its respective class. The class with the highest probability across all classifiers is chosen as the predicted label.

In Multinomial Logistic Regression, also known as Softmax Regression, a single model is trained to predict the probabilities for all classes at once. The Softmax function is used to ensure that the probabilities of all possible classes sum to 1.

For example, if you're classifying images of digits 0-9 (a 10-class problem), the model will output 10 probabilities, one for each digit. The softmax function assigns a probability to each class (e.g., 0.1 for digit 3, 0.05 for digit 7, etc.), and the class with the highest probability is predicted.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, auc
from sklearn.preprocessing import label_binarize

# Binarize the output (convert multiclass labels into one-hot encoding)
n_classes = len(np.unique(y_test))
y_test_bin = label_binarize(y_test, classes=np.arange(n_classes))
y_score = model.predict_proba(x_test)  # Probability estimates for each class

# Compute ROC curve and AUC for each class
plt.figure(figsize=(8,6))
for i in range(n_classes):
    fpr, tpr, _ = roc_curve(y_test_bin[:, i], y_score[:, i])
    roc_auc = auc(fpr, tpr)
    plt.plot(fpr, tpr, label=f'Class {i} (AUC = {roc_auc:.2f})')

# Random classifier line
plt.plot([0, 1], [0, 1], 'k--', label="Random Classifier")

plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Multiclass ROC Curve')
plt.legend()
plt.show()
```
